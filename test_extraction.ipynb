## Extraire les coordonn√©es des pixels

### ouvrir le fichier txt et recuperer les pixels positions

```python
import re

# Chemin du fichier texte
fichier_texte = "contained_segments_50.txt"

# Liste pour stocker toutes les positions des pixels
positions_pixels = []

# Ouvrir le fichier en mode lecture
with open(fichier_texte, 'r') as fichier:
    # Lire chaque ligne du fichier
    for ligne in fichier:
        # Utiliser une expression r√©guli√®re pour extraire le champ "Pixels"
        match_pixels = re.search(r'Pixels: (\[\[.*\]\])', ligne)
        
        # Si une correspondance est trouv√©e, ajouter les positions des pixels √† la liste
        if match_pixels:
            positions_pixels.append(eval(match_pixels.group(1)))

# Afficher les positions des pixels
# for position in positions_pixels:
#     print("Positions des pixels :", position)
positions_pixels
```

### Convertir en coordonn√©es spatiales

```python
import math

def meters_to_longitude(distance_in_meters, latitude):
    degrees_longitude = distance_in_meters / (111000.0 * math.cos(math.radians(latitude)))
    return degrees_longitude

def pixels_to_coordinates(pixel, top_left_lat, top_left_lon, resolution_lat):
    x, y = pixel
    latitude = top_left_lat - (y * resolution_lat)
    degrees_longitude = (x * 10) / (111000.0 * math.cos(math.radians(latitude)))
    longitude = top_left_lon + degrees_longitude
    return latitude, longitude

# Coordonn√©es du coin sup√©rieur gauche
top_left_latitude = 11.37
top_left_longitude = -3.89

# R√©solution spatiale
resolution_lat = 1 / 111000

# Appliquer la fonction √† chaque paire de coordonn√©es
coordonnees_geographiques = []
for positions in positions_pixels:
    for pixel_coordinates in positions:
        latitude, longitude = pixels_to_coordinates(pixel_coordinates, top_left_latitude, top_left_longitude, resolution_lat)
        coordonnees_geographiques.append((latitude, longitude))

# Afficher les coordonn√©es g√©ographiques
# for i, coordonnees in enumerate(coordonnees_geographiques, start=1):
#     print(f"Coordonn√©es g√©ographiques pour le pixel {i}: Latitude {coordonnees[0]}, Longitude {coordonnees[1]}")
```

```python
from pyproj import CRS, Transformer, TransformerGroup, areas_of_use

# Ouvrir le fichier raster
path = 'first_acquisition.tif'
with rasterio.open(path) as src:
    # Obtenir l'EPSG code du syst√®me de coordonn√©es
    src_epsg = src.crs.to_epsg()

    # D√©finir le syst√®me de coordonn√©es de destination (WGS84)
    dst_crs = CRS.from_epsg(4326)  # WGS84

    # Coordonn√©es des pixels √† convertir
    pixel_coordinates = (1145, 2921)

    # Obtenir les coordonn√©es g√©ographiques du coin sup√©rieur gauche de la zone d'√©tude
    top_left_lat = 11.37
    top_left_lon = -3.89

    # D√©finir manuellement la zone d'utilisation en fonction de votre zone d'√©tude
    area_of_use = areas_of_use("Burkina Faso", top_left_lon, 10.96, top_left_lat, -3.42)

    # D√©finir les syst√®mes de coordonn√©es source et destination avec area_of_use
    src_crs = CRS.from_epsg(src_epsg, area_of_use=area_of_use)
    dst_crs = CRS.from_epsg(4326, area_of_use=area_of_use)

    # Utiliser un Transformer pour effectuer la conversion
    transformer = Transformer.from_crs(src_crs, dst_crs)

    # Convertir les coordonn√©es
    lon, lat = transformer.transform(pixel_coordinates[0], pixel_coordinates[1])

    # Ajuster les coordonn√©es en fonction du coin sup√©rieur gauche de la zone d'√©tude
    lat = top_left_lat - lat
    lon = top_left_lon + lon

    print(f'Latitude: {lat}, Longitude: {lon}')
```
